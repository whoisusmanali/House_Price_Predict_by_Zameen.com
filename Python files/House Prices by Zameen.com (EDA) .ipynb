```python
df=pd.read_csv('/content/EDA_dataset.csv')
df.head()
```

```python
df.drop(['Unnamed: 0'],axis=1,inplace=True)
df.head()
```

```python
#let's see how which city is more expensive to buy or rent house
fig=px.bar(df.groupby('city',as_index=False)['price'].mean().sort_values(by='price',ascending=False).head(10),
           x='city',y='price',color='city',labels={'city':'City Name','price':'Price'},
           template='ggplot2',text='price',title='<b> Average Price by City')
fig.show()
```

### Insights
1. Lahore is more expensive with PKR 25M to buy property. <br>
2. Karachi is on second number with PKR 20M.<br>
3. In islamabad you just have PKR 13M to buy a good property.<br>
4. Rawalpindi and Faisalabad are on last with PKR 9M and 8.1M respectively.<br>

```python
#Price by Property type
fig=px.pie(df.groupby('property_type',as_index=False)['price'].mean().sort_values(by='price',ascending=False).head(10),
           names='property_type',values='price',color='property_type',
           labels={'property_type':'Property Type ','price':'Average Price'},template='ggplot2',
           title='<b>Average price by Property type')
fig.update_layout(title_x=0.5,legend=dict(orientation='h',yanchor='bottom',
                                          y=1.02,xanchor='right',x=1))
```

### Insights
1. Firstly, Farm Houses are more expensive in Pakistan with PKR 60M with is the total 52.8% of the total ads.<br>
2. To buy House is less expensive than Farm houses with PKR 25M with is also 21.9% of the total ads.<br>
3. Remainng are the Flat, upper portion and lower portion are the on the least position and take remaing percentage.



```python
#Which city has more ad posting
px.funnel(df.groupby('city',as_index=False)['price'].count().sort_values(by='price',
                    ascending=False).head(15),y='city',x='price',
          color_discrete_sequence=['orange'],labels={'price':'count'},
          template='seaborn',title='<b> Which city has more ad posting')
```

### Insights
1. Faislabad is less in this list with only 6892 ads while Karachi tops the list with 54.65k ads.
2. Lahore and Islamabad are near to each other with 38.49k and 34.45k respectively.
3. Rawalpindi posted 18.94k ads.

```python
#Top 20 agent contributers
fig=px.bar(df.groupby('agent',as_index=False)['price'].count().sort_values(by='price',ascending=False).head(20)[1:22],
           x='agent',y='price',color='agent',labels={'agent':'Agent Name','price':'Total contributions'},
           template='presentation',text='price',title='<b> Top 20 agent Contributer')
fig.show()
```

```python
#Top 20 agency contributers
fig=px.bar(df.groupby('agency',as_index=False)['price'].count().sort_values(by='price',ascending=False).head(20)[1:22],
           x='agency',y='price',color='agency',labels={'agency':'agency Name','price':'Total contributions'},
           template='plotly_dark',text='price',title='<b> Top 20 agency Contributer')
fig.show()
```
